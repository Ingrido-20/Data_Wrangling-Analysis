Time-Series Analysis: Resample the cleaned data to an hourly frequency. Calculate a 24-hour rolling average for the primary metric (e.g., Pressure or Temperature). 

```python
clean_df = clean_df.sort_values("timestamp")

clean_df = clean_df.set_index("timestamp")

hourly = clean_df.resample("1H").mean(numeric_only=True)

hourly.head()
hourly["Pressure_24hr_Rolling"] = (
    hourly["Pressure_PSI"]
    .rolling(window=24)
    .mean()
)

hourly.head()

```
